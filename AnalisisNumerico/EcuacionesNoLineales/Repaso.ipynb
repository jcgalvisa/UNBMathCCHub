# Comparación asintótica

```julia
# using Pkg
# Pkg.add("Plots")
using Plots
```

Considere $\{x_n\}$ y $\{\alpha_n\}$ tale que $x_n\to 0$ and $\alpha_n\to 0$.

Decimos que $x_n=\mathcal{O}(\alpha_n)$ si existe $N_0$ y $C$ tal que $|{x_n}| \leq |{\alpha_n}|$ para todo $n\geq N_0$.

Quiere decir que la sucesión   $\{\frac{x_n}{\alpha_n}\}$ permanece acotada (cuando $\alpha_n\not=0$). También que   $\{x_n\}$ converge a  $0$ al menos tan rápido como  $\{\alpha_n\}$.

Adicionalmente, $x_n=\mathcal{o}(\alpha_n)$, significa  que $\displaystyle \lim_{n\to\infty} \frac{x_n}{\alpha_n}=0$. Quiere decir que   $\{x_n\}$  converge a  $0$    más rápido que  $\{\alpha_n\}$.


Ejemplo: $\frac{n+1}{n^2}=\mathcal{O}\left(\frac{1}{n}\right)$. Esto puede ser demostrado de forma simple. Podemos ilustrar esta afirmación con una gráfica de de los los cocientes de los valores de las dos sucesiones.

```julia
n=1:50
x= (n.+1)./n.^2;
α= 1 ./n;
scatter(n,x./α,lw=1,label="x(n)/α(n)")
```

Ejemplo: $\frac{1}{n\log(n)}=\mathcal{o}\left(\frac{1}{n}\right)$. Esto puede ser demostrado de forma simple. Podemos ilustrar esta afirmación con una gráfica de de los los cocientes de los valores de las dos sucesiones.

```julia
n=1:100
x= 1 ./(n.*log.(n));
α= 1 ./n;
scatter(n,x./α,lw=5,label="x(n)/α(n)",ms=4)
```

Ejemplo: $e^{-n}=\mathcal{o}\left(\frac{1}{n^k}\right)$ for $k=1,2,\dots$. Esto puede ser demostrado de forma simple. Podemos ilustrar esta afirmación con una gráfica de de los los cocientes de los valores de las dos sucesiones.

```julia
n=1:50
x= exp.(-n);
α= 1 ./n.^4;
scatter(n,x./α,lw=5,label="x_n",ms=5)
```

Ejemplo: $\sin(x)=x-\frac{x^3}{6}+\mathcal{O}(x^5)$ cuando $x\to0$ (puede pensar que se coman sucesiones $x_n\to 0$. Esto puede ser verificado fácilemente. Para una ilustración pomdemos gráficar la funcion $\sin(x)$ junto a la función $\frac{x^3}{6}-\mathcal{O}(x^5)$ la función $\sin(x)=x-\frac{x^3}{6}+\mathcal{O}(x^5)$ junto con la función $x^5/5!$.

```julia
x=-2:0.1:2
f=sin.(x)
g=x-x.^3 ./6
plot(x,f, lw=4,label="sin(x)")
plot!(x,g,lw=4, label="x-x^3/6",legend=:bottomright)
```

```julia
x=-0.2:0.01:0.2
f=sin.(x)
g=x-x.^3 ./6
plot(x,f-g, lw=4,label="sin(x)-x-x^3/6")
scatter!(x,x.^5/120, label="x^5/5!")
```

Decimos que $f(x)=\mathcal{O}(g(x))$ cuando $x\to \infty $ si para $x$ grande tenemos $|f(x)|\leq C |g(x)|$ for some positive constant $C$. De forma similar decimos que  $f(x)=\mathcal{O}(g(x))$ cuando $x\to a $ si existen $C$ y $r$ tales que $|f(x)| \leq C|g(x)|$ para todo $x$ con $|x-a|<r$. Se denota $|f(x)|\preceq |g(x)|$ para $x\approx a$.

Ejemplo: $\sqrt{x^2+1}=\mathcal{O}(x)$. Esto puede ser demostrado de forma simple. Podemos ilustrar esta afirmación con una gráfica de las dos funciones.  

```julia
x=0:50:1000
f=sqrt.( x.^2 .+1)
g=x
plot(x,f)
scatter!(x,g)
```

Decimos que $f(x)=\mathcal{o}(g(x))$ cuando $x\to a$ si $f(x)/g(x)\to 0$. Recuerede que $f(x)\to 0$ y $g(x)\to 0$. 

# Velocidad de convergencia

Considere ${x_n}$ una sucesion con limite $a$, es decir, $x_n\to a$. 

La convergencia es (por lo menos) lineal si existen $0< \alpha <1$  y $N$ tal  que ${|x_{n+1}-a|}\leq \alpha {|x_{n+1}-a|}$ para todo $n\geq N$. Informalmente hablando quiere decir que "necesitamos hacer (en promedio) siempre el mismo esfuerzo para ganar una determinada cantidad de decimales correctos".

Ejemplo: La sucesión $x_n=0.5^n$ es un ejemplo típico de convergencia lineal. 

```julia
n=30;
ndc=0;
x=zeros(n,1)
for n=1:n
    x[n]=0.5^n;
    if x[n]<10.0^(-ndc-1)
        ndc=ndc+1;
        println("....Un decimal correcto mas")
    end    
    println("x[n]=",x[n])
    
end
```

```julia
scatter(1:n,-log.(x), label="-log(x)")
```

La convergencia es superlineal si existe $\epsilon_n\to 0$ tal que $|x_{n+1}-a| \leq \epsilon_n |x_n-a|$. Intuitivamente es cada vez es más fácil (menos iteraciones) ganar cierta cantidad de decimales correctos!

Ejemplo: $x_n= 0.5^{n^2}$. Podemos estimar los valores $\epsilon_n$ and ilustar este caso. 

```julia
n=20
x=zeros(n,1)
ϵ=zeros(n,1)
x[1]=0.5;
for i=2:n
    x[i]=0.5^(i^2)
    ϵ[i-1]=abs(x[i]-0.0)/abs(x[i-1]-0.0)
end
scatter(ϵ)
```

```julia
scatter(log.(x))
```

La convergencia es (al menos) cuadrática si existen $C$ y $N$ tales que para $n\geq N$ tenemos $|x_{n+1}-a|\leq C|x_n-a|^2$. Intuitivamente podemos decir que  asintóticamente, necesitamos hacer (en promedio) siempre el mismo esfuerzo para duplicar el número de  decimales correctos!!!

En general la convergencia es de orden $\alpha$ si para $n$ grande tenemos  $|x_{n+1}-a|\leq C|x_n-a|^\alpha$.

Ejemplo: $x_n= 0.5^{2^n}$.

```julia
n=1:15
x= 0.5 .^(2 .^n);
scatter(n,log.(x),lw=4,label="log(x(n))")
```

# Expansión de Taylor

Suponga que $f\in C^n([a,b])$ y que $f^{n+1}$ existe en $(a,b)$. Entonces para todo $c,x\in[a,b]$ existe $\xi$ entre $c$ y $x$ tal que 
$$
f(x)=f(c)+f'(c)(x-c)+\\+\frac{1}{2} f''(c)(x-c)^2+\cdots+\frac{1}{n!}f^{(n)}(c)(x-c)^n+E_n(x)$$
donde 
$$
E_n(x)=\frac{1}{(n+1)!}f^{(n+1)}(\zeta)(x-c)^{n+1}.$$

Suponga que $f\in C^{n+1}([a,b])$. Entonces para todo $c,x\in[a,b]$ vale 
$$
f(x)=f(c)+f'(c)(x-c)+\\+\frac{1}{2} f''(c)(x-c)^2+\cdots+\frac{1}{n!}f^{(n)}(c)(x-c)^n+R_n(x)$$
donde 
$$
R_n(x)=\frac{1}{n!}\int_c^x f^{(n+1)}(t)(x-t)^ndt.$$
